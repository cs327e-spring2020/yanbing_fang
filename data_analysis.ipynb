### The daily increase of confirmed cases in Texas

```python
%%bigquery
with cte as(select ROW_NUMBER() over(ORDER BY date) as date_rank, date, sum(confirmed) as confirmed,
            FROM covid_19_staging.cases_SQL_Final_Date
            where state = "Texas" and country = "US"
           group by date)
SELECT c1.*, c1.confirmed - c2.confirmed  as daily_growth
FROM cte c1
LEFT JOIN cte c2 ON c2.date_rank = c1.date_rank - 1
order by daily_growth desc limit 5
```

```python
%%bigquery
select daily_growth_rate from (
with cte as(select ROW_NUMBER() over(ORDER BY date) as date_rank, date, sum(confirmed) as confirmed,
            FROM covid_19_staging.cases_SQL_Final_Date
            where state = "Texas" and country = "US"
           group by date)
SELECT c1.*, round((c1.confirmed - c2.confirmed)*100/c2.confirmed,2)  as daily_growth_rate
FROM cte c1
LEFT JOIN cte c2 ON c2.date_rank = c1.date_rank - 1)
where date = "2020-04-09"
```

Therefore, on April 9th Texas has the most daily growth, with a growth rate of 30.16%

### The daily increase of confirmed cases in China

```python
%%bigquery
create table covid_19_staging.cases_in_China as
select t2.state, t1.date_rank, t2.date,t2.confirmed from(
select ROW_NUMBER() over(ORDER BY date) as date_rank, date
from covid_19_staging.cases_SQL_Final_Date
where country in ('Mainland China', 'China')
group by date) t1
right join 
(select state, date, confirmed from covid_19_staging.cases_SQL_Final_Date
where country in ('Mainland China', 'China')) t2
on t1.date = t2.date 
order by date
```

```python
%%bigquery
select * from covid_19_staging.cases_in_China limit 6
```

```python
%%bigquery
WITH cte AS(SELECT state, date_rank, date, confirmed,
            FROM covid_19_staging.cases_in_China )
SELECT c1.*, c1.confirmed - c2.confirmed as daily_growth
FROM cte c1
LEFT JOIN cte c2 ON c1.state = c2.state AND c2.date_rank = c1.date_rank - 1
limit 50
```

```python
%%bigquery
select state, date, confirmed,daily_growth
from (WITH cte AS(SELECT state, date_rank, date, confirmed,
            FROM covid_19_staging.cases_in_China )
SELECT c1.*, c1.confirmed - c2.confirmed as daily_growth
FROM cte c1
LEFT JOIN cte c2 ON c1.state = c2.state AND c2.date_rank = c1.date_rank - 1)
where state = "Hubei" order by daily_growth desc limit 10
```

Therefore, on Feb 13th Hubei province has the most daily increase

```python
%%bigquery
select date, sum(daily_growth) as daily_growth
from (
WITH cte AS(SELECT state, date_rank, date, confirmed,
            FROM covid_19_staging.cases_in_China )
SELECT c1.*, c1.confirmed - c2.confirmed as daily_growth
FROM cte c1
LEFT JOIN cte c2 ON c1.state = c2.state AND c2.date_rank = c1.date_rank - 1
)
group by date
order by daily_growth desc
```

Therefore, Feb 13th also has the most increased cases nationalwide

```python
%%bigquery
with cte as(select ROW_NUMBER() over(ORDER BY date) as date_rank, date, sum(confirmed) as confirmed,
            FROM covid_19_staging.cases_in_China
           group by date)
SELECT c1.*, round((c1.confirmed - c2.confirmed)*100/c2.confirmed,2)  as daily_growth_rate
FROM cte c1
LEFT JOIN cte c2 ON c2.date_rank = c1.date_rank - 1
```
